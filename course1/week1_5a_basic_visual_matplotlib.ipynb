## Visualization Tools

In this video, we will try some of the visualization toolkits in Python. Exploratory data analysis is important in Bayesian statistics, and more generally about all data science projects, because visualizion tools are crucial to understand data. So in the last lecture of the first week, we will discuss how to use the Python Matplotlib library to visually inspect the data, using tools you may be already familiar with, such as histograms and scatter plots, and we will do so using the Jupyter notebook as our environment.

Nowadays Matplotlib is a well-established data visualization library that is supported in various environments such as in Python scripts, in the iPython shell, as well as in graphical user interface toolkits such as the Jupyter notebook. For those who don't know what the Jupyter notebook is, it's an open source web application that allows you to create, share and present contents in live coding demonstrations, that's exactly the form of presentation that we are giving coding lectures that you are watching.

There is a specialized subject called exploratory data analysis to describe the techniques to explain the patterns and valuable insights given some data.
In this video, we will learn how to use Matplotlib to create plots that describes probability distribution of variables, as well as the relationship between different variables, 


```
# Just as we use the np shorthand for numpy, we will use some standard shorthands for 
# importing visualization tools

import numpy as np
# Jupyter has some specialized support for Matplotlib and so if you start a Jupyter notebook, 
# all you need to do is import Matplotlib with the scripting interface
# Let's first import the scripting interface as plt
import matplotlib.pyplot as plt
```

```
# we will learn how to use Matplotlib to create plots that describes probability distribution of variables, 
# We'll first learn how to visualize the relationship between different variables
# Let us create a variable x to store a sequence of 100 values starting from negative pi to pi

x = np.linspace(-np.pi, np.pi, 100)

# Let's plot the sine and cosine function between negative pi to pi. We can use the plt.plot function,
# and plug the x and y axis in the first and second arguments which signify the x and y coordinates.

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

# So I get my sine and cosine functions, but does not look informative
```

```
# We can customize a bit by adding a horizontal and vertical line passing through the origin.
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

# To add a horizontal line, we can call the plt.axhline function and pass a color parameter specifying what 
# color line you are going to show
plt.axhline(color = 'black')
# To create a vertical line, on the other hand, we'd use the plt.axvline function and pass the same color specification
plt.axvline(color = 'black')
# Cool! But we still want a plot title as well as the axis labels
```

```
# So we copy the same code, now I add a label on each plot by passing the third parameter
# Note that the label will render latex so I can enclose the math functions with the dollar sign $.
plt.plot(x, np.sin(x), label = '$sin(x)$')
plt.plot(x, np.cos(x), label = '$cos(x)$')
plt.axhline(color = 'black')
plt.axvline(color = 'black')

# We can customize the legend by calling the plt.legend function. In this case, we put the label at the upper right corner
plt.legend(loc="upper right")

# Finally we can add a title or x-axis and y-axis labels, then just write 
# plt.title for title, plt.xlabel for x label, plt.ylabel for y label
# and enter it as a string in single quotes
plt.title('Sine and Cosine function from $-\pi$ to $\pi$')
plt.xlabel('x')
plt.ylabel('$sin(x)$ and $cos(x)$')
# That looks much better! That's enough to pretty much plot anything straightforward. 
```

Matplotlib can be used to create histograms. A histogram shows the frequency on the vertical axis and the horizontal axis is another dimension. Usually it has bins, where every bin has a minimum and maximum value. Let's plot some data.

```
# Let's import pandas as pd
import pandas as pd
# and use the convenient function read_csv to take in the IQ dataset
iq = pd.read_csv('assets/iq.csv')

# Now let's take a glimpse on the first ten rows
iq.head(10)
# The IQ test was used by Binet to evaluate basic intellectual functions such as reasoning and problem-solving skills in school children.
# A high IQ score indicates better reasoning and problem-solving abilities than average individuals and may signal intellectual potential. 
```

```
# Let's use the plt.hist method in matplotlib to create a very basic histogram about IQ distribution
# First we set the number of bins to 20
num_bins = 20

# We pass through the IQ column in the data, and by using the bins function, matplotlib will automatically
# create 20 evenly spaced bins.
plt.hist(iq['IQ'], bins = num_bins)
# and finally, label the x and y axes, give the title and plot the histogram
plt.xlabel('IQ')
plt.ylabel('Frequency')
plt.title('Histogram of IQ Distribution')

# Now we get a pretty simple histogram. What this shows is the distribution of IQ among 10000 test takers.
# I personally find it a bit difficult to read these without edge and face colors for each bin because they all just kind of run together.
# facecolor='blue', alpha=0.5
```

```
# So let's add in some edge color so that's more clear

# We can add those in by going back to the plt.hist and pass the edgecolor and facecolor arguments 
# Personally I like to set the transparency level using the alpha at alpha=0.5 so that the edge colors are clear to see
plt.hist(iq['IQ'], bins = num_bins, edgecolor = 'black', facecolor = 'blue', alpha = 0.5)
plt.xlabel('IQ')
plt.ylabel('Frequency')
plt.title('Histogram of IQ Distribution')

# Now the graph becomes a bit more clear
# We wanted our data broken up into 20 equally-spaced bins, so the histogram calculated those IQs for us. 
# Let's see the x-axis that the first bin looks like between somewhere 50 something, and last bin looks like somewhere slightly less than 160. 
# You will see the bin with the highest frequency locates at somewhere between 93 to 98; there are around 1500 people here scored 93 to 98 points in the IQ test. 
# It's interesting to see that the closer to the highest frequency bin, the more people's IQ score will fall into that bin, meaning the more 
# likely that a individual will score that IQ range. Cool!
```

A potentially confusing feature of Matplotlib is its dual interfaces: a convenient MATLAB-style interface, and an object-oriented interface. 
Matplotlib was originally developed as a Python alternative for MATLAB users, so many syntaxes and MATLAB-style tools are composed in the pyplot interface. Let's plot some data to see the differences.

```
# Let's import pandas as pd
import pandas as pd
# and use the convenient function read_csv to take in the admission dataset
admission = pd.read_csv('assets/admission.csv')
# Let's look at the first few rows
admission.head()
```

The dataset contains several parameters which are considered important during the application of UCLA master's programs. The dataset was originally created to predict the chance of admission from an Indian perspective, and serve the purpose of providing a fair idea about their chances for a particular university given their profiles.
For each observation, the parameters include 1) GRE scores where the maximum score is 340, 2) TOEFL scores with a maximum of 120 and the chance of admit ranging from 0 to 1. 
From an educational perspective, it's interesting to understand how a student's chance of admission is influenced by student's language profiency and aptitude represented in standardized test such as GRE and TOEFL score.

```
# Let's play around with plt functions to examine the relationship between 1) gre_score and prob_admit, and 2) toefl_score and prob_admit
# First we create a plot figure
plt.figure()

# We create a subplot with two rows and one column
# Now we use the first panel
plt.subplot(2, 1, 1)
# that shows the change of admission probability based on applicants' gre_score
# We can also add g. as the third parameter to indicate
# the format of the points to use, which in this case a green dot.
plt.plot(admission['gre_score'], admission['prob_admit'], 'g.')

# Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes it is nice to have
# finer control. A convenient way to adjust axis limits is to use the plt.xlim() and plt.ylim() methods
#plt.ylim(-1.1, 1.1)
#plt.legend(bbox_to_anchor=(1.35,1), borderaxespad = 0.)
plt.xlabel("GRE Score")
plt.ylabel("Chance of Admit")

# and use the second panel
plt.subplot(2, 1, 2)
# to show the change of admission probability based on applicants' toefl_score
plt.plot(admission['toefl_score'], admission['prob_admit'], 'b.')
# Again adjust the y-axis limit to increase visibility
#plt.ylim(-1.1, 1.1)
# Add a legend at the right hand side
#plt.legend(bbox_to_anchor=(1.35,1), borderaxespad = 0)
plt.xlabel('TOEFL Score')
plt.ylabel("Chance of Admit")
plt.suptitle("How does a student's GRE and TOEFL score affect the chance of admission?")

# scatterplot
# positive linear relationship 
```

To have more control on the figure, rather than depending on the notion of "active" figure and axes, we can rely on the object-oriented interface, which enables the use of plotting functions as methods for figure and axes objects.
When creating more complicated plots, knowing how to specify and call objects in object-oriented approach can simplify the work.

In object-oriented interface, most plt functions translates directly to ax methods (i.e. plt.plot() becomes ax.plot(), plt.legend() translates to ax.legend()), 
but some have modifications. In particular, we'll see how functions to set limits, labels and titles are modified when transitioning between MATLAB-style and object-oriented methods as follows:

* `plt.xlabel()` $\rightarrow$ `ax.set_xlabel()`
* `plt.ylabel()` $\rightarrow$ `ax.set_ylabel()`
* `plt.xlim()` $\rightarrow$ `ax.set_xlim()`
* `plt.ylim()` $\rightarrow$ `ax.set_ylim()`
* `plt.title()` $\rightarrow$ `ax.set_title()`

It's very convenient though if you use the `ax.set()` method to set all properties for plots at once, rather than calling these functions individually.

```
# So again we start with creating a grid of plots, but store the information in variables figure and axes
fig, ax = plt.subplots(2)

# and we call plot() method on the appropriate object
ax[0].plot(admission['gre_score'], admission['prob_admit'], 'g.')
ax[1].plot(admission['toefl_score'], admission['prob_admit'], 'b.')
# then you can use .set function which wraps up all necessary arguments to customize the plot. In this case, we
# set the x and y limits, x and y labels respectively.
ax[0].set(xlabel = "GRE Score", ylabel = "Chance of Admit")
ax[1].set(xlabel = 'TOEFL Score', ylabel = "Chance of Admit")
plt.suptitle("How does a student's GRE and TOEFL score affect the chance of admission?")
```
