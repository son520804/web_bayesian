## Matplotlib Object Oriented Interface

In Matplotlib “Object Oriented Interface”, every figure is divided into some objects and the object hierarchy should be clearly described. A Figure in matplotlib is divided into two different objects.
* Figure object
* Axes object

* A **Figure object** can contain one or more axes objects. One axes represents one plot inside figure. In this tutorial we will be working with the axes object directly for all kinds of plotting.

* One or more **axes** objects are required to get us started with plotting. There are more then one ways an axes object can be obtained.

```python
import matplotlib.pyplot as plt
fig = plt.figure()

# Generate a grid of 2x2 subplots and get
# axes object for 1st location
ax1 = fig.add_subplot(2,2,1)
ax1.set_title('First Location')

# Get the axes object for subplot at 2nd 
# location
ax2 = fig.add_subplot(2,2,2)
ax2.set_title('Second Location')

# Get the axes object for subplot at 3rd 
# location
ax3 = fig.add_subplot(2,2,3)
ax3.set_xlabel('Third Location')

# Get the axes object for subplot at 4th 
# location
ax4 = fig.add_subplot(2,2,4)
ax4.set_xlabel('Fourth Location')

# Display
plt.show()
```

* plot(x, y) : Generate y vs x graph
* set_xlabel() : Label for the X-axis
* set_ylabel() : Label for the Y-axis
* set_title() : Title of the plot
* legend() : Generate legend for the graph
* hist() : Generate histogram plot

and so on.

## Bar graph

```python
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
```

```python
x = np.arange(len(labels))
x
```

```python
width = 0.4
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x-width/2, men_means, width, label="Men")
rects2 = ax.bar(x+width/2, women_means, width, label="Women")
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel("Scores")
ax.set_title("Scores by Group and Gender")
plt.gcf().set_size_inches(12,7)
plt.show()
```

## how to draw multiple plots using for loop (not subplots)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

greenhouse = pd.read_csv("week1.csv").set_index("country_or_area")
greenhouse
```

```python
for i in list(greenhouse.columns):
    plt.figure(figsize=(8,5))
    plt.bar(greenhouse.index, greenhouse[i])
    plt.title(str(i))
    plt.xticks(rotation = "vertical")
    plt.show()
```
